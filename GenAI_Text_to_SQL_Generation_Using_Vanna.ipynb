```python
# Import library
import vanna as vn
from vanna.remote import VannaDefault
```

### What are the top 10 albums by sales?

```python
# Initialize VannaDefault with specified model and API key
vanna = VannaDefault(model='chinook', api_key=vn.get_api_key('bayuworkspace28@gmail.com'))
# Connect to SQLite database using the provided URL
vanna.connect_to_sqlite('https://vanna.ai/Chinook.sqlite')
# Ask a query to retrieve the top 10 albums by sales
vanna.ask("What are the top 10 albums by sales?")
```
